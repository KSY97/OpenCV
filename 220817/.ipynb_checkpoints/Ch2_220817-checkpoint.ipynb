```python
img_files = glob.glob('../fig/my_images/*.*')

idx = 0
cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

while True:
    img = cv2.imread(img_files[idx])
    
    if img is None:
        print('failed')
        break
    
    cv2.imshow('scene', img)
    
    if cv2.waitKey(1000) == 27:
        break
    
    idx += 1
    if idx >= len(img_files):
        idx = 0

cv2.destroyAllWindows()
```

```python
src = cv2.imread('../fig/cow.png')
dst = cv2.imread('../fig/green.png')

src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

retval, mask = cv2.threshold(src_gray, 240, 255, cv2.THRESH_BINARY_INV)

h, w = dst.shape[:2]
src_h, src_w = src.shape[:2]

dst_copy = dst[(h//2)-(src_h//2)+100:(h//2)+(src_h//2)+100 , (w//2)-(src_w//2):(w//2)+(src_w//2)]

dst_resize = cv2.resize(dst, (w, h))

cv2.copyTo(src, mask, dst_copy)

cv2.imshow('mask', mask)
cv2.imshow('dst_c', dst_copy)
cv2.imshow('dst', dst)

cv2.imwrite('copy_cow.jpg', dst)

while True:
    key = cv2.waitKey()
    if key == 27 or key == ord('q'):
        break
        
cv2.destroyAllWindows()
```

```python
img = cv2.imread('../fig/cat.bmp', 0)

img_origin = img.copy()

if img is None:
    print('failed')
    sys.exit()
    
cv2.imshow('img', img)
    
while True:
    key = cv2.waitKey()
    
    if key == ord('i'):
        img = 255-img
        cv2.imshow('img', img)
    elif key == ord('e'):
        img = cv2.Canny(img, 150, 150)
        cv2.imshow('img', img)
    elif key == ord('r'):
        img = img_origin.copy()
        cv2.imshow('img', img)
    elif key == ord('s'):
        cv2.imwrite('img.png', img)
    elif key == 27:
        break
    
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('failed')
    sys.exit()

w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) 
h = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
fps = int(cap.get(cv2.CAP_PROP_FPS))
fourcc = cv2.VideoWriter_fourcc(*'DIVX')
edge = False
flip = False


while True:
    ret, frame = cap.read()
    
    if not ret:
        print('failed')
        break
    
    
    if edge is True:
        frame =  cv2.Canny(frame, 70, 80)
    
    if flip is True:
        frame = cv2.flip(frame, 1)
        
        
    cv2.imshow('img', frame)
        
    
        
    key = cv2.waitKey(20)
    
    if key == 27:
        break
    elif key == ord('e'):
        edge = not edge
    elif key == ord('f'):
        flip = not flip
    
cap.release()
cv2.destroyAllWindows()
```

```python
def call_trackbar(pos):
    img[:] = np.round(img_copy * (pos/255), 0).astype(np.uint8)
    cv2.imshow('img', img)

img = cv2.imread('../fig/cat.bmp', 0)
img_copy = img.copy()

if img is None:
    print('failed')
    sys.exit()
    
cv2.imshow('img', img)
cv2.createTrackbar('level', 'img', 0, 255, call_trackbar)

while True:
    key = cv2.waitKey()
    if key == 27:
        break
cv2.destroyAllWindows()
```
